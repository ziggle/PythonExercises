```python
import pandas as pd
df = pd.read_csv("TSLA.csv")
df = df.head(11).sort_values(by='Date')
df = df.astype({"Open":'float',
                "High":'float',
                "Low":'float',
                "Close":'float',
                "Adj Close":'float',
                "Volume":'float'})
df['Gain'] = df['Close'] - df['Open']
remarkable_filter = (df['Volume'] > 30000000) | (df['Gain'] > 0)
df['Remarkable'] = False
df.loc[remarkable_filter, ['Remarkable']] = True
```

```python
import pandas as pd
df = pd.read_csv('Happiness/world-happiness-report-2019.csv')
df
```

```python
df['first_five_Letter'] = df['Country (region)'].str.extract(r'(^\w{5})')
df
```

```python
x=df[df['Country (region)'].str.count('^[pP].*')>0]
x
```

```python
x=df[df['Country (region)'].str.match('^P.*')== True]
x
```

```python
x=df[df['Country (region)'].str.contains('^I.*')==True]
x
```

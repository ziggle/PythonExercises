```python
from sklearn.datasets import load_wine
import pandas as pd
boston = load_wine()
boston_df = pd.DataFrame(boston.data, columns=boston.feature_names)
boston_df['TARGET'] = pd.Series(boston.target)
boston_df.head()
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
boston_df['malic_acid'].plot()
```

```python
boston_df['malic_acid'].plot.hist()
```

```python
boston_df['malic_acid'].plot.hist(title='Proportion of owner-occupied units built prior to 1940')
```

```python
boston_df['malic_acid'].plot.hist(title='Proportion of owner-occupied units built prior to 1940', colormap='jet')
plt.xlabel('malic_acid')
plt.ylabel('malic_acid')
```

```python
correlations = boston_df.corr()
correlations = correlations['TARGET']
print(correlations)
```

```python
boston_df.plot.scatter(x='flavanoids', y='TARGET')
```

```python
boston_df.groupby('CHAS').boxplot(fontsize=20,rot=90,figsize=(20,10),patch_artist=True)
```

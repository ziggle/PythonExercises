```python
#import the necessary libraries:
import pandas as pd
import numpy as np
import seaborn as sns

df = pd.read_csv('heart.csv')
df
```

```python
sns.boxplot( x=df["slp"], y=df["thalachh"] )
```

```python
Group_0 = df.loc[df['slp'] == 0]
Group_1 = df.loc[df['slp'] == 1]
Group_2 = df.loc[df['slp'] == 2]
```

```python
#Get the len for the three groups:
print('Len of Group 0 is: ', len(Group_0))
print('Len of Group 1 is: ', len(Group_1))
print('Len of Group 2 is: ', len(Group_2))
```

```python
#Get mean and sd for 'thalachh' in the three groups:
import numpy as np
print('Mean thalachh in Group 0 is: ', np.mean(Group_0['thalachh']))
print('Mean thalachh in Group 1 is: ', np.mean(Group_1['thalachh']))
print('Mean thalachh in Group 2 is: ', np.mean(Group_2['thalachh']))
```

```python
from scipy.stats import f_oneway

statistic, p_value = f_oneway(Group_0['thalachh'], Group_1['thalachh'], Group_2['thalachh'])
print('ANOVA F valus is: ', statistic, '\np-value = ', p_value)
```

```python
import seaborn as sns
penguins=sns.load_dataset("penguins")
sns.displot(data=penguins,x="flipper_length_mm")
```

```python
df=sns.load_dataset("titanic")
sns.countplot(x=df["class"])
```

```python
import seaborn as sns
penguins=sns.load_dataset("penguins")
sns.jointplot(data=penguins,x="bill_length_mm",y="bill_depth_mm")
```

```python
sns.boxplot(data=df,x="age",y="class")
```

```python
sns.barplot(data=penguins,x="island",y="body_mass_g",hue='sex')
```

```python
flights=sns.load_dataset("flights")
sns.lineplot(data=flights,x="year",y="passengers",hue="month",style="month")
```

```python
sns.pairplot(penguins,hue="species",markers=["o","s","D"])
```

```python
sns.violinplot(data=df,x="deck",y="age",hue="alive",split=True)
```

```python
tips=sns.load_dataset("tips")
sns.stripplot(data=tips,x="total_bill",y="day",hue="day",legend=False)
```

```python
sns.lmplot(
    data=penguins, x="bill_length_mm",y="bill_depth_mm",
    col="species",row="sex",height=3,
    facet_kws=dict(sharex=False,sharey=False)
)
```

```python
glue=sns.load_dataset("glue")
ax=sns.heatmap(glue)
ax.set(xlabel="",ylabel="")
ax.xaxis.tick_top()
```

```python
g=sns.PairGrid(penguins)
g.map(sns.scatterplot)
```

```python
g=sns.FacetGrid(tips,col="time",row="sex")
g.map_dataframe(sns.histplot,x="total_bill")
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
df = sns.load_dataset('penguins')
sns.regplot(data=df, x='flipper_length_mm', y='body_mass_g')
plt.show()
```

```python
df = sns.load_dataset('titanic')
sns.lmplot(x='fare', y='age', hue='class', data=df)
# Show the plot
plt.show()
```

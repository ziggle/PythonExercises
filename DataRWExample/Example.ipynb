```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_table("data.txt")
display(df)
```

```python
df=pd.read_table("data2.txt")
display(df)
```

```python
df=pd.read_table("data2.txt",sep=",")
display(df)
```

```python
df=pd.read_table("data2.txt",sep=",",header=None)
display(df)
```

```python
df=pd.read_table("data2.txt",sep=",",header=None,names=["name","score","sex"])
display(df)
```

```python
df=pd.read_table("data2.txt",sep=",",header=None,names=["name","score","sex"],index_col="name")
display(df)
```

```python
df2=pd.read_table("data3.txt", sep=",")
display(df2)
```

```python
df2=pd.read_table("data3.txt", sep=",", index_col=["lesson","name"])
display(df2)
```

```python
df3=pd.read_table("data4.txt", sep=",")
display(df3)
```

```python
df3=pd.read_table("data4.txt", sep=",",skiprows=[0,2])
display(df3)
```

```python
df3=pd.read_table("data4.txt", sep=",",skiprows=[0,2],usecols=[0,1])
display(df3)
```

```python
df3=pd.read_table("data4.txt", sep=",",skiprows=[0,2],usecols=[0,1],nrows=3)
display(df3)
```

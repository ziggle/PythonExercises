```python
import matplotlib.pyplot as plt

# Sample data
categories = ['Category A', 'Category B', 'Category C', 'Category D']
values = [25, 40, 15, 30]

# Creating a bar chart with customized colors and widths
plt.bar(categories, values, color=['aquamarine', 'limegreen', 'tomato', 'violet'], width=0.6)
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Customized Bar Colors and Widths')
plt.xticks(rotation=45)
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Sample data
data = sns.load_dataset('iris')

# Creating a basic box plot using Seaborn
sns.boxplot(x='species', y='sepal_length', data=data)
plt.xlabel('Species')
plt.ylabel('Sepal Length')
plt.title('Basic Box Plot with Seaborn')
plt.show()
```

```python
import plotly.express as px

# Sample data
data = px.data.tips()

# Create an interactive scatter plot
fig = px.scatter(data, x="total_bill", y="tip", color="sex", size="size",
                 title="Interactive Scatter Plot using Plotly")
fig.show()
```

```python
import plotly.express as px
import pandas as pd
heart=px.data.gapminder()
#print(heart)


# Compare Heart Attack vs Sex 
df = pd.crosstab(heart['country'],heart['continent'])

sns.set_style("white")

df.plot(kind="bar",
        figsize=(6,6),
        color=['#c64343', '#e1d3c1']);

plt.title("Heart Attack Risk vs Sex ", fontsize=16)
plt.xlabel("0 = Lower Risk                  1 = Higher Risk", fontsize=16)
plt.ylabel("Amount", fontsize=16)
plt.legend(["Female","Male"], fontsize=14)
plt.xticks(rotation=0)
```

```python
from bokeh.plotting import figure, output_file, show

# Sample data
x = [1, 2, 3, 4, 5]
y = [3, 5, 8, 6, 7]

# Create a scatter plot
p = figure(title="Scatter Plot Example using Bokeh")
p.circle(x, y, size=10, color="navy")

output_file("scatter.html")
show(p)
```

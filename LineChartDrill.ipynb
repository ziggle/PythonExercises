```python
import seaborn as sns
import matplotlib.pyplot as plt

# Sample data
x = [1, 2, 3, 4, 5]
y = [3, 5, 8, 6, 7]

# Create a line plot using Seaborn
sns.lineplot(x=x, y=y, marker='o', color='b', label='Line Chart with Seaborn')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Line Chart with Seaborn')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt

# Sample data
x = [1, 2, 3, 4, 5]
y = [3, 5, 8, 6, 7]

# Create a line chart using Matplotlib
plt.plot(x, y, marker='o', linestyle='-', color='b', label='Line Chart with Matplotlib')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Line Chart with Matplotlib')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt

# Sample data
x = [1, 2, 3, 4, 5]
y = [3, 5, 8, 6, 7]

# Creating a line chart with customized line style and color
plt.plot(x, y, marker='o', linestyle='--', color='purple', label='Customized Line')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Customized Line Style and Color')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt

# Sample data for multiple lines
y1 = [2, 4, 6, 8, 10]
y2 = [1, 3, 5, 7, 9]

# Creating a line chart with multiple lines
plt.plot(x, y1, marker='o', linestyle='-', color='blue', label='Line 1')
plt.plot(x, y2, marker='s', linestyle='--', color='red', label='Line 2')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Multiple Lines on a Single Line Chart')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt

# Creating a line chart with markers and annotations
plt.plot(x, y, marker='o', linestyle='-', color='green', label='Line with Markers')
for i, j in zip(x, y):
    plt.annotate(f'({i}, {j})', (i, j), textcoords="offset points", xytext=(0,10), ha='center')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Line Chart with Markers and Annotations')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import pandas as pd

# Sample time series data
dates = pd.date_range('2023-01-01', periods=5)
values = [10, 20, 15, 30, 25]

# Create a line chart with time series data
plt.plot(dates, values, marker='o', linestyle='-', color='purple', label='Time Series')
plt.xlabel('Date')
plt.ylabel('Value')
plt.title('Time Series Line Chart')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Sample data for multiple time series
values1 = [12, 18, 14, 25, 20]
values2 = [8, 15, 10, 22, 18]

# Creating a line chart with multiple time series
plt.plot(dates, values1, marker='o', linestyle='-', color='green', label='Series 1')
plt.plot(dates, values2, marker='s', linestyle='--', color='orange', label='Series 2')
plt.xlabel('Date')
plt.ylabel('Value')
plt.title('Multiple Time Series Lines')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings("ignore")
```

```python
df_prod = pd.read_csv('./SourceFiles/Classic Models - Products.csv',sep=",")
df_ord = pd.read_csv('./SourceFiles/Classic Models - OrderDetails.csv',sep=";")
```

```python
df_ord.head()
```

```python
df_prod.head()
```

```python
sales_all_features = pd.merge(df_ord,df_prod, how='inner', on='productCode') 
```

```python
sales_scope=sales_all_features[["orderNumber","productLine","productCode","quantityOrdered","priceEach"]]
sales_scope.head()
```

```python
sales_scope['OrderValue']=sales_scope['quantityOrdered'] * sales_scope['priceEach']
```

```python
sales_scope.tail()
```

```python
sales_stats = sales_scope.groupby(by="productLine")["OrderValue"].sum().sort_values(ascending=False)
display(sales_stats)
```

```python
sales_stats.to_csv('OutputDataFrame1.csv',sep=";")
```

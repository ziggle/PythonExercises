```python
def check(stringy):
    if stringy.__contains__("/"):
        print("here")
        zig = pd.to_datetime(stringy, format="mixed").strftime('%m/%d/%Y')
    elif stringy.__contains__("-"):
        print("here1")
        zig = pd.to_datetime(stringy,format="mixed",dayfirst=True).strftime('%m/%d/%Y')
    else:
        print("here2")
        zig = pd.to_datetime(stringy,format="mixed").strftime('%m/%d/%Y')
    return zig

check("8-Mar-18")
```

```python
import pandas as pd
import numpy as np
df = pd.read_csv("train.csv")
df.head()
```

```python
df.shape
```

```python
df.groupby("brand")[['price','volume']].mean()
```

```python
df.groupby("brand").mean(numeric_only=True)
```

```python
df.groupby(['brand','model']).agg({'volume':['sum']})
```

```python
df.groupby(['brand','model']).agg({'volume':[np.sum]})
```

```python
df.groupby(['brand','model']).agg({'price':[np.mean],'volume':[np.sum]})
```

```python
df.groupby(['brand','model']).apply(lambda x:np.mean(x['price'])-x['price'])
```

```python
avg_price_dict=df.groupby(['brand','model']).agg({'price':['mean']})
avg_price_dict.columns=['brand_model_avg_price']
avg_price_dict.reset_index(inplace=True)
df_new=pd.merge(df,avg_price_dict,on=['brand','model'],how='left')
avg_price_dict
```

```python
df['brand_model_avg_price']=df.groupby(['brand','model'])['price'].transform('mean')
```

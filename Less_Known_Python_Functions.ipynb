```python
import pandas as pd
df=pd.DataFrame({"cola":[1.27, -0.61,1.09,0.159, 1.80, -0.28],
                 "colb":[1,3,-3,-5,-2,-4],
                 "colc":[0.08,0.42,0.41,0.043, 0.25, 0.11]})
display(df)
```

```python
display(df.clip(lower=0))
display(df)
```

```python
df.clip(lower=0, upper=1)
```

```python
df.eval("cola = cola * 10")
```

```python
df.eval("new = cola * colb")
```

```python
import numpy as np
df1=pd.DataFrame({"cola":[1,3,4,6,7],"colb":[3,3,3,4,1],"colc":["a","b","d","g","f"]})
df2=pd.DataFrame({"cola":[0.0,4.0,np.nan,2.0,8.0],
                  "colb":[1.0,np.nan,4.0,2.0,1.0],
                  "colc":[np.nan,"b","k",np.nan,"f"]})
```

```python
display(df1)
display(df2)
```

```python
df2.combine_first(df1)
```

```python
df3=pd.DataFrame({"cola":[719,281,360,551,345],
                 "colb":[6,7,2,7,9],
                 "colc":[45,42,70,64,35]})
```

```python
display(df3)
```

```python
df3.transform(lambda x:np.log(x))
```

```python
df3.transform([lambda x: np.log(x), np.sqrt])
```

```python
df3['diff_c'] = df3['colc'].diff()
df3
```

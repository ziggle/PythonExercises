```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("cars.csv")
print(data)
```

```python
# Get the shape of the data
print(data.shape)

# Get the column names
print(data.columns)

# Get the data types of each column
print(data.dtypes)

# Get the first few rows of the data
print(data.head())

# Get the last few rows of the data
print(data.tail())

# Get the summary statistics of the data
print(data.describe())
```

```python
# Check for missing values
print(data.isnull().sum())
display(data)

# Fill missing values with a constant value
#data.fillna(0, inplace=True)

# Drop rows with missing values
data.dropna(inplace=True)

# Fill missing values with the mean value of the column
#data.fillna(data.mean(), inplace=True)

# Fill missing values with the median value of the column
#data.fillna(data.median(), inplace=True)
```

```python
# Fill missing values with a constant value
data.select_dtypes(include=np.number).fillna(0, inplace=True)

# Drop rows with missing values
data.select_dtypes(include=np.number).dropna(inplace=True)

# Fill missing values with the mean value of the column
data.select_dtypes(include=np.number).fillna(data.select_dtypes(include=np.number).mean(), inplace=True)

# Fill missing values with the median value of the column
data.select_dtypes(include=np.number).fillna(data.select_dtypes(include=np.number).median(), inplace=True)
```

```python
# Check for duplicate rows
print(data.duplicated().sum())

# Drop duplicate rows
data.drop_duplicates(inplace=True)
```

```python
# Convert all the values in a column to lowercase
data["Make"] = data["Make"].str.lower()

# Replace all occurrences of a value with another value
data.replace("bmw", "BMW", inplace=True)

# Rename columns
data.rename(columns={"Model": "Car_Model", "Price": "Sale_Price"}, inplace=True)

# Drop columns
#data.drop(columns=["Unnamed: 0"], inplace=True)
```

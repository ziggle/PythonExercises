```python
# Factorisation in Python
from sympy import symbols, factor

x, y = symbols("x y")

expr = (x - 2) * (x - 2) - (y - 3) * (y - 3)
print(factor(expr))
# (x - y + 1)*(x + y - 5)
```

```python
# Expansion in Python
from sympy import Symbol, expand

x = Symbol("x")
y = Symbol("y")

expr = (x - y + 1)*(x + y - 5)
expr = expand(expr)
print(expr)
# x**2 - 4*x - y**2 + 6*y - 5
```

```python
# Expressions Solving
from sympy import Symbol

x = Symbol("x")
y = Symbol("y")
expr = x**2 - 2*x - y**3 + 7*y - 3

result = expr.subs({x: 1, y: 2})
print(result)
# 2
```

```python
# Expressions Calculation
from sympy import Symbol

x = Symbol("x")
y = Symbol("y")

expr1 = x * y + y * y
expr2 = x * x + x * y

print(expr1 + expr2)
# x**2 + 2*x*y + y**2

print(expr1 - expr2)
# -x**2 + y**2
```

```python
# Solve single-variable equations
from sympy import symbols, solve

x, y = symbols('x, y')
expr = x * x - 5 * x + 6
print(solve(expr))
# [2, 3]


print(solve(expr, dict=True))
# [{x: 2}, {x: 3}]
```

```python
# Solve two-variable equations
from sympy import symbols, solve

x, y = symbols('x, y')
expr1 = 3 * x + 4 * y - 34
expr2 = 2 * x + 3 * y - 56

print(solve((expr1, expr2), dict=True))
# [{x: -122, y: 100}]
```

```python
from sympy import symbols, sin, N

x, y = symbols("x y")
expr = sin(x)

print(expr.subs(x, 0))
# 0
print(expr.subs(x, 3))
# sin(3)
print(N(expr.subs(x, 3)))
# 0.141120008059867
```

```python
from sympy import symbols, sin, cos, tan, sec, trigsimp

x, y = symbols("x y")

print(trigsimp(sin(x) ** 2 + cos(x) ** 2))
# 1
print(trigsimp(cos(x) * tan(x)))
# sin(x)
```

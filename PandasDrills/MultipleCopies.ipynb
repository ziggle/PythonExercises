```python
import glob
import pandas as pd
```

```python
path = './Data'
files = glob.glob(path + "/*.csv")
files
```

```python
df_list = (pd.read_csv(file) for file in files)
```

```python
df = pd.concat(df_list, ignore_index=True)
df.head()
```

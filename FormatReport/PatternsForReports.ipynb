```python
import pandas as pd
def flatten_cols(df_):
    cols = ['_'.join(map(str,vals)) for vals in df_.columns.to_flat_index()]
    df_.columns=cols
    return df_

sales=pd.read_csv("data.csv", parse_dates=['Date'])
sales.head()
```

```python
def checker(df_):
    print(df_)
    print()
    return df_
```

```python
(sales
.query("Status=='Paid'")
.groupby([pd.Grouper(key='Date',freq='MS'),'Item'])
.agg(total_sales=('Value','sum'),
     total_items=('Value','count'))
.pipe(checker)
.unstack()
.pipe(checker)
.pipe(flatten_cols)
)
```

```python
import numpy as np
np.random.seed(10)
data = np.random.normal(200, 10, 2000)
```

```python
import matplotlib.pyplot as plt
fig = plt.figure(figsize =(10, 7))
plt.boxplot(data)
plt.title("BoxPlot")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
df = sns.load_dataset('penguins')
fig = plt.figure(figsize = (6, 5))
ax = sns.boxplot( y=df["bill_depth_mm"] );
ax.set(title="Box Plot via Seaborn")
plt.show()
```

```python
ax = sns.boxplot( y=df["island"], x=df["bill_depth_mm"])
ax.set(title="Box Plot via Seaborn")
plt.show()
```

```python
import plotly.express as px
fig = px.box(df, y="bill_depth_mm")
fig.show()
```

```python
fig = px.box(df, x="island", y="bill_depth_mm")
fig.show()
```

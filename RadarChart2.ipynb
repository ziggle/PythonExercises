```python
import numpy as np
import matplotlib.pyplot as plt

# Data
categories = ['A', 'B', 'C', 'D', 'E']
values = [4,3,4,6,2]

# Plot
fig = plt.figure()
ax = fig.add_subplot(111, polar=True)
ax.plot(np.linspace(0, 2 * np.pi, len(categories), endpoint=False), values)
ax.fill(np.linspace(0, 2 * np.pi, len(categories), endpoint=False), values, alpha=0.25)

# Set ticks
ax.set_xticks(np.linspace(0, 2 * np.pi, len(categories), endpoint=False))
ax.set_xticklabels(categories)

# Title
plt.title("Radar Chart")
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Data
categories = ['A', 'B', 'C', 'D', 'E']
values1 = [5, 3, 5, 2, 1]
values2 = [3, 4, 3, 5, 4]

# Plot
fig = plt.figure()
ax = fig.add_subplot(111, polar=True)

# The first variable
ax.plot(np.linspace(0, 2 * np.pi, len(categories), endpoint=False), values1)
ax.fill(np.linspace(0, 2 * np.pi, len(categories), endpoint=False), values1, alpha=0.25)

# The second variable
ax.plot(np.linspace(0, 2 * np.pi, len(categories), endpoint=False), values2)
ax.fill(np.linspace(0, 2 * np.pi, len(categories), endpoint=False), values2, alpha=0.25)

# Set ticks
ax.set_xticks(np.linspace(0, 2 * np.pi, len(categories), endpoint=False))
ax.set_xticklabels(categories)

# Legend
ax.legend(['Variable 1', 'Variable 2'], loc='best')

# Title
plt.title("Radar Chart")
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Categories
indicators = ['Survival Rate', 'Recurrence Rate', 'Quality of Life', 'Treatment Effectiveness', 'Side Effects']

# Data
values1 = [0.8, 0.5, 0.7, 0.6, 0.4]
values2 = [0.6, 0.7, 0.8, 0.5, 0.6]

# Plot
fig = plt.figure()
ax = fig.add_subplot(111, polar=True)

# The first variable
ax.plot(np.linspace(0, 2 * np.pi, len(indicators), endpoint=False), values1)
ax.fill(np.linspace(0, 2 * np.pi, len(indicators), endpoint=False), values1, alpha=0.25)

# The second variable
ax.plot(np.linspace(0, 2 * np.pi, len(indicators), endpoint=False), values2)
ax.fill(np.linspace(0, 2 * np.pi, len(indicators), endpoint=False), values2, alpha=0.25)

# Set ticks and format
ax.set_xticks(np.linspace(0, 2 * np.pi, len(indicators), endpoint=False))
ax.set_xticklabels(indicators)
ax.tick_params(axis='x', labelsize=12, rotation=0, labelcolor='red',pad=5)

# Legend
legend = plt.legend(['Variable 1', 'Variable 2'], loc='best')

# Title
plt.title("Radar Chart")
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Categories
indicators = ['Survival Rate', 'Recurrence Rate', 'Quality of Life', 'Treatment Effectiveness', 'Side Effects']

# Data
values1 = [0.8, 0.5, 0.7, 0.6, 0.4]
values2 = [0.6, 0.7, 0.8, 0.5, 0.6]

# Plot
fig = plt.figure()
ax = fig.add_subplot(111, polar=True)

# The first variable
ax.plot(np.linspace(0, 2 * np.pi, len(indicators), endpoint=False), values1)
ax.fill(np.linspace(0, 2 * np.pi, len(indicators), endpoint=False), values1, alpha=0.25)

# The second variable
ax.plot(np.linspace(0, 2 * np.pi, len(indicators), endpoint=False), values2)
ax.fill(np.linspace(0, 2 * np.pi, len(indicators), endpoint=False), values2, alpha=0.25)

# Set x ticks and format
ax.set_xticks(np.linspace(0, 2 * np.pi, len(indicators), endpoint=False))
ax.set_xticklabels(indicators)
ax.tick_params(axis='x', labelsize=12, rotation=0, labelcolor='red',pad=5)

# Legend
legend = plt.legend(['Variable 1', 'Variable 2'], loc='best')

# Title
plt.title("Radar Chart")

# Set y ticks
ticks = np.linspace(0, 1, 6)  # The distances of scale
ax.set_yticks(ticks)

# Set data label format
ax.set_yticklabels(['{:.0%}'.format(tick) for tick in ticks])  # Use p
```

```python
# Import the required library
import matplotlib.pyplot as plt
plt.style.use("ggplot")

# Create some data
payment_type = ["Check", "Credit Card", "Debit Card", "UPI", "Cash"]
average_transac = [20, 46, 27, 70, 10]

# Create the bar chart
fig, ax = plt.subplots()
ax.bar(payment_type, average_transac)

# Customize
ax.set(title= 'Bar Chart Example', 
       xlabel= 'Type of Transaction', 
       ylabel= 'Average Amount($)')
ax.set_yticks([15, 30, 45, 60, 75])

# Display the chart
fig.savefig("Bar_plot.png")
plt.show()
```

```python
# Import the required library
import matplotlib.pyplot as plt
plt.style.use("bmh")

# Create some data
years = ['1950-59', '1960-69', '1970-79', '1980-89', '1990-99', '2000-09',
         '2010-19', '2020-21']
temp_aus = [13.9, 13.9, 14.0, 14.2, 14.0, 14.4, 14.9, 14.4] 
temp_north_america = [12.5, 12.1, 12.0, 12.5, 12.9, 13.1, 13.6, 14.4]
temp_asia = [15.6, 15.7, 15.6, 15.7, 16.2, 16.4, 16.7, 17.4]

# Create the bar chart
fig, ax = plt.subplots()
ax.plot(years, temp_aus, color = 'purple', marker = 'o', label = 'Australia')
ax.plot(years, temp_asia, color = 'green', marker = 'o', label = 'Asia')
ax.plot(years, temp_north_america, color = 'red', marker = 'o', label = 'North America')

# Customize
ax.set(title='Line Chart Example', xlabel='Years', ylabel='Average temperatures(C)')
ax.set_ylim(11,18)
ax.legend(loc = 'lower right')
plt.grid(True)

# Display the chart
fig.savefig("Line_Chart.png")
plt.show()
```

```python
# Import the required library
import numpy as np
import matplotlib.pyplot as plt
plt.style.use("fivethirtyeight")

# Create some data
N = 50
price = np.random.randint(0,50,N)
sales = np.random.randint(0,100,N)
colors = np.random.rand(N)
area = (20 * np.random.rand(N))**2  

# Create the bar chart
fig, ax = plt.subplots()
ax.scatter(price, sales, s=area, c=colors, alpha=0.85)

# Customize
ax.set(title='Scatter Plot Example', xlabel='price ($)', ylabel='Sales')
plt.grid(True)

# Display the chart
fig.savefig("Scatter_plot.png")
plt.show()
```

```python
# Import the required library
import matplotlib.pyplot as plt
import numpy as np
plt.style.use("classic")

# Create some data
data = np.random.normal(size=1000)

# Create the histogram
fig, ax = plt.subplots()
ax.hist(data, bins=20, color = 'lime')

# Customize
ax.set(title='Histogram Example', xlabel='Value', ylabel='Frequency')

# Display the chart
fig.savefig("Histogram.png")
plt.show()
```

```python
# Import the required library
import matplotlib.pyplot as plt
import numpy as np
plt.style.use("seaborn-poster")

# Create some data
region = ["North", "South", "East", "West"]
revenue = [400, 150, 220, 350]

# Create the Pie-plot
fig, ax = plt.subplots()
ax.pie(revenue, labels = region, autopct='%1.2f%%', startangle=70)

# Customize
ax.set(title='Pie Chart')
ax.legend(title = "Regions:", loc = 'lower right')

# Display the chart
fig.savefig("Pie.png")
plt.show()
```

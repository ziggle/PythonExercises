This is an exercise in simulating the results from A/D converter and HPM10

```python
from numpy import random
vBAT=3.7
adc = 0.005
#mn=0.332_750_00
#mx=0.332_968_75
mn=0.327312
mx=0.333018
rmn=1/mx
rmx = 1/mn
rng = mx-mn
print(f'min is {mn:.4f}, max is {mx:.4f}')
print(f'range is {rng/2:0.4f}')
ratio = random.uniform(mn, mx,size=10000)
#ratio = random.normal(0.332668,0.00011397,1000)
lsad=random.uniform(-adc,adc,10000)
#vbat  = random.uniform(vBAT*0.327312,vBAT*0.333018,10000)
vha = (vBAT*ratio)+lsad
print(vha.min(),vha.max(),vha.mean(),vha.max()-vha.mean(),vha.min()-vha.mean())
print(vBAT*rng/2+0.005)
print(3.7*rng/2)

```

```python
import pandas as pd
df = pd.DataFrame( {"Customer Country":["PR", "PR","PR", "United States","United States"],
                    "Customer FName":["Sean","Carol","Mary","Mary","Victoria"],
                    "Customer Id":[568,3341,7459,7459,8100],
                    "Customer Segment":["Consumer","Consumer","Consumer","Consumer","Corporate"],
                    "Customer State":["PR","PR","PR","CA","MN"]

})
```

```python
df.loc[:, 'Customer Country':'Customer State']
```

```python
df.loc[[0,1,2], 'Customer Country':'Customer State']
```

```python
df.iloc[[0,1,2,3] , [1,3,4]]
```

```python
df.iloc[[0,1,2,3] , 2:4]
```

```python
df.iloc[:, [0,2,3]]
```

```python
df.iloc[:,[0,2,3]]
```

```python
df[['Customer Country']]
```

```python
df[['Customer Country', 'Customer State']]
```

```python
df.loc[df['Customer Id'] > 1000]
```

```python
df.loc[df['Customer Country'] == 'United States']
```

```python
df.iloc[[0, 2, 4]]
```

```python
df[df['Customer Country'].isin(['United States', 'Puerto Rico'])]
```

```python
df[~df['Customer Country'].isin(['United States'])]
```

```python
df.rename(columns={"Customer Fname" : "Customer_Fname"}, inplace=True)
```

```python
df.query('`Customer Id` > 4000')
```

```python
df.query('`Customer Id` > 4000 and `Customer State` == "MN"')
```

```python
df[df['Customer Id'].between(4000, 8000)]
```

```python
df[df['Customer Country'].str.startswith('United')]
```

```python
df[df['Customer Country'].str.contains('P')]
```

```python
df[df['Customer Country'].str.endswith('States')]
```

```python
df.loc[df['Customer FName'].str.contains('^M'), :]
```

```python
df.loc[(df['Customer State']=='PR')&(df['Customer FName'].str.contains('^M')), :]
```

```python
df.loc[df['Customer FName'].str.contains('.*ar'), :]
```

```python
import pandas as pd
crimes=pd.read_csv("data.csv")
crimes.head()
```

```python
(crimes
  .sample(500)
  .assign(month_name=lambda df:pd.to_datetime(df['Date']).dt.month_name())
  .pipe(lambda df:pd.crosstab(index=df['Primary Type'],columns=df.month_name))
)
```

```python
(crimes
  .sample(500)
  .assign(month_name=lambda df:pd.to_datetime(df['Date']).dt.month_name()
  .astype('category')
  .cat.reorder_categories(['January','February','March','April','May','June',
                           'July','August','September','October','November','December'],ordered=True))
  .pipe(lambda df:pd.crosstab(index=df['Primary Type'],columns=df.month_name))
)
```

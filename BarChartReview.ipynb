```python
import matplotlib.pyplot as plt

# Sample data
categories = ['Category A', 'Category B', 'Category C', 'Category D']
values = [25, 40, 15, 30]

# Creating a bar chart with customized colors and widths
plt.bar(categories, values, color=['aquamarine', 'limegreen', 'tomato', 'violet'], width=0.6)
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Customized Bar Colors and Widths')
plt.xticks(rotation=45)
plt.show()
```

```python
# Adding data labels to the bars
for i, value in enumerate(values):
    plt.text(i, value + 1, str(value), ha='center', va='bottom', fontsize=10)

plt.bar(categories, values, color='cadetblue')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Bar Chart with Data Labels')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.bar(categories, values, color='skyblue', width=0.5, edgecolor='black', label='Customized Bars')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Customized Bar Chart')
plt.legend()
plt.xticks(rotation=45)
plt.show()
```

```python
import seaborn as sns
#!pip install seaborn --upgrade
# Creating a horizontal bar chart using Seaborn
plt.figure(figsize=(8, 6))
sns.barplot(y=categories, x=values, color='orange')
plt.xlabel('Values')
plt.ylabel('Categories')
plt.title('Horizontal Bar Chart with Seaborn')
plt.show()
```

```python
plt.bar(categories, values, color='orchid')
plt.xlabel('Categories', fontsize=12, fontweight='bold')
plt.ylabel('Values', fontsize=12, fontweight='bold')
plt.title('Customized Axes Labels and Title', fontsize=14, fontweight='bold')
plt.xticks(rotation=45)
plt.show()
```

```python
# Customizing colors and style
import pandas as pd
colors = ['lightcoral', 'lightgreen']
sns.set_palette(colors)
data = pd.DataFrame({"Category":["Category A","Category B","Category C","Category D"],"Type 1":[25,40,15,30],"Type 2":[35,20,25,10]})


plt.figure(figsize=(10, 6))
sns.barplot(x='Category', y='value', hue='variable', data=pd.melt(data,"Category"))
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Customized Grouped Bar Chart with Seaborn')
plt.xticks(rotation=45)
plt.legend(title='Type')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x='Category', 
            y='value', 
            hue='variable', 
            data=pd.melt(data,"Category"),
            )
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Grouped Bar Chart with Custom Bar Widths')
plt.xticks(rotation=45)
plt.legend(title='Type')
plt.show()
```

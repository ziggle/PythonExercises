```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('./pokedex.csv',index_col=0)
df.head()
```

```python
cols_to_keep=['pokedex_number', 'name', 'generation',
       'species',
       'type_1', 'type_2', 'height_m', 'weight_kg',
       'abilities_number', 'ability_1', 'ability_2', 
       'total_points', 'hp', 'attack', 'defense', 'sp_attack', 'sp_defense',
       'speed', 
       'percentage_male', 'against_normal', 'against_fire',
       'against_water', 'against_electric', 'against_grass', 'against_ice',
       'against_fight', 'against_poison', 'against_ground', 'against_flying',
       'against_psychic', 'against_bug', 'against_rock', 'against_ghost',
       'against_dragon', 'against_dark', 'against_steel', 'against_fairy']
```

```python
df =df[cols_to_keep]
```

```python
df.shape
```

```python
df.head()
```

```python
pd.crosstab(index=df.species,columns=df.generation)
```

```python
pd.crosstab(index=df.species,columns=df.generation,margins=True,margins_name='Total')
```

```python
pd.crosstab(index=df.species,columns=df.generation,margins=True,margins_name='Total',normalize=True)
```

```python
pd.crosstab(index=df.species,columns=df.generation,margins=True,margins_name='Total',normalize='index')
```

```python
pd.crosstab(index=df.species,columns=df.generation,margins=True,margins_name='Total',normalize='columns')
```

```python
pd.crosstab(index=df.species,columns=df.generation,margins=True,values=df.weight_kg,aggfunc='max')
```

```python
pd.crosstab(index=df.species,columns=df.generation,rownames=['which kind?'],colnames=['gen'])
```

```python
pd.crosstab(index=df.species,columns=df.generation,values=df.percentage_male,aggfunc='mean',dropna=False)
```

```python
pd.crosstab(index=df.species,columns=df.generation,values=df.percentage_male,aggfunc='mean',dropna=False).fillna(0)
```

```python
pd.crosstab(index=[df.generation,df.species],columns=df.abilities_number,values=df.total_points,aggfunc='median').fillna(0)
```

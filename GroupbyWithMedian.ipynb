```python
import pandas as pd
stackoverflowDF=pd.read_csv("survey_results_public.csv",low_memory=False)
display(stackoverflowDF)
print(stackoverflowDF.Salary)
```

```python
stackoverflowDF.info()
stackoverflowDF.filter(items=['Salary','Age','CurrencySymbol'])
```

```python
groupedDF = stackoverflowDF.filter(items=['Country',
                                          'Age'])  # Remove columns
groupedDF = groupedDF.groupby(['Age']).count()  # Perform Aggregate
print(groupedDF)
```

```python
groupedDF = stackoverflowDF.filter(items=['Salary','Age','CurrencySymbol'])  # Remove columns
groupedDF = groupedDF.loc[groupedDF['CurrencySymbol'] == 'USD']
groupedDF.sort_values(by=['Salary'],inplace=True,ascending=False)  # Sort descending
groupedDF = groupedDF.groupby(['Age'])['Salary'].median()
#groupedDF.set_index(["Age"])
#groupedDF = groupedDF[['Age','Salary']].median()
#print(list(groupedDF))
#groupedDF.sort_values(by=['Salary'],inplace=True,ascending=False)  # Sort descending

print(groupedDF)
```

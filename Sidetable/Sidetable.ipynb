```python
import pandas as pd
df=pd.read_csv("WA_Fn-UseC_-HR-Employee-Attrition.csv")
df.info()
```

```python
df.stb.freq(["Department"])
```

```python
df.stb.freq(["Department","BusinessTravel"])
```

```python
df.stb.freq(["Department"], value='MonthlyIncome')
```

```python
df.stb.freq(["Department"], value='MonthlyIncome',style=True)
```

```python
df.stb.counts()
```

```python
df.stb.counts(exclude='number')
```

```python
df.stb.missing(style=True)
```

```python
df.stb.missing(clip_0=True,style=True)
```

```python
df[['Attrition','Department','Gender']].groupby(['Gender','Department']).agg('count').stb.subtotal()
```

```python
df[['Attrition','Department','Gender']].groupby(['Gender','Department']).agg({"Attrition":'count'}).stb.subtotal()
```

```python
df[['Attrition','Department','Gender']].groupby(['Gender','Gender','Department'])\
                                        .agg({"Attrition":'count'}).stb.subtotal(sub_level=[2])
```

```python
stacked_df=df[['Attrition','Department','Gender']].groupby(['Gender','Gender','Department'])\
                                        .agg({"Attrition":'count'}).stb.subtotal(sub_level=[2])
```

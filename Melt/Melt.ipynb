```python
import pandas as pd
# The top four rows are not required so they are removed. Delimiter is ','
df= pd.read_csv('API_NY.GDP.MKTP.CD_DS2_en_csv_v2_6548046.csv',skiprows=4,sep=',')
# Only keep the US data
df_usa = df[df['Country Code']=='USA']
df_USA=df_usa.copy()
```

```python
display(df_USA)
```

```python
# drop columns without any values
df_USA.dropna(axis='columns',inplace=True)
#drop two columns 
df_USA.drop(['Indicator Name','Indicator Code'],axis=1,inplace=True)
# suppress scientic notation and remove all decimal points
pd.options.display.float_format='{:,.0f}'.format
```

```python
df_USA
```

```python
df_USA = pd.melt(df_USA, id_vars=['Country Name','Country Code'],
                         value_vars=['2018','2019','2021','2021','2022'],
                         var_name = 'Year', value_name = 'GDP')
```

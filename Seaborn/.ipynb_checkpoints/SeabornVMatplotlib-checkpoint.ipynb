```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
plt.style.use('classic')
%matplotlib inline
```

```python
rng = np.random.RandomState(0)
x = np.linspace(0,10,250)
y =np.cumsum(rng.randn(250,6),0)
```

```python
plt.plot(x,y)
plt.legend("ABCDEF",loc='best',ncol=2)
```

```python
import seaborn as sns
sns.set()
```

```python
plt.plot(x,y)
plt.legend("ABCDEF",loc='upper left',ncol=2)
```

```python
iris=sns.load_dataset('iris')
iris.head()
```

```python
setosa = iris.loc[iris.species=='setosa']
virginia = iris.loc[iris.species=='virginica']
setosa.sepal_length.plot.hist()
                 
```

```python
sns.kdeplot(
setosa.sepal_length,shade=True,color='r'
)
```

```python
sns.histplot(iris.sepal_length,kde=True)
```

```python
sns.histplot(
iris.sepal_length,kde=True,color='r'
)
sns.histplot(
iris.sepal_width,kde=True,color='b'
)
```

```python
sns.kdeplot(
data=iris,x="sepal_length",y="sepal_width"
)
```

```python
sns.kdeplot(
data=iris,x="sepal_length",y="sepal_width",
    shade=True
)
```

```python
with sns.axes_style('white'):
    sns.jointplot(
    data=iris,x='sepal_length',
    y='sepal_width',kind='kde')
```

```python
with sns.axes_style('white'):
    sns.jointplot(
    data=iris,x='sepal_length',
    y='sepal_width',kind='hex')
```

```python
sns.pairplot(iris,hue='species')
```

```python
tips=sns.load_dataset('tips')
tips['tip_percent']=tips['tip']*100/tips['total_bill']
grid=sns.FacetGrid(
tips,row='smoker',col='time',margin_titles=True)
grid.map(
plt.hist,"tip_percent",bins=np.linspace(0,40,15))
```

```python
with sns.axes_style(style='ticks'):
    g = sns.catplot(x='day',y='total_bill',data=tips,kind='box')
g.set_axis_labels('Day','Total Account')
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='reg')
```

```python
with sns.axes_style('white'):
    g = sns.catplot(x='size',data=tips,aspect=2,kind='count',color='steelblue')
    g.set_xticklabels(step=5)
```

```python
import matplotlib
from matplotlib import dates
from matplotlib.colors import LinearSegmentedColormap
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
from matplotlib import rcParams
```

```python
df = pd.read_excel('https://www.nahb.org/~/media/NAHB/news-and-economics/docs/housing-economics/hmi/202207/t2-national-hmi-history-202207.xls',index_col=[0],skiprows=[0,1]) 
df
```

```python
def plot_hmi(df):
    for font in ['xtick.color','ytick.color','axes.labelcolor']:
        matplotlib.rcParams[font] = "#ffffff"
    matplotlib.rcParams['font.family'] = 'Roboto'
    matplotlib.rcParams['axes.titlepad'] = 40
    
    fig = plt.Figure(figsize=(8,6), dpi=200)
    fig.patch.set_facecolor('#1a1a1a')
    ax = fig.add_axes((0.2,0.4,.8,0.4))
    cmap = LinearSegmentedColormap.from_list('my_gradient', (
    (0.000, (0.612, 0.824, 0.898)),
    (0.147, (0.027, 0.686, 0.922)),
    (0.272, (0.282, 0.694, 0.459)),
    (0.434, (0.471, 0.753, 0.196)),
    (0.727, (1.000, 0.776, 0.125)),
    (1.000, (0.851, 0.173, 0.239))))
    
    ax.bar(df.date, df.HMI-50, width=31, color=cmap(df.HMI/df.HMI.max()), zorder=10)
    
    ax.set_yticks(range(-50,51,10))
    ax.set_yticklabels(range(0,101,10))
    
    years = dates.YearLocator(base=10)
    ax.xaxis.set_major_locator(years)
    ax.xaxis.set_minor_locator(dates.YearLocator(base=5))
    ax.grid(b=True, which='major',linewidth=1, zorder=0, color='#555')
    ax.grid(b=True, which='minor',linewidth=0.5, zorder=0, color='#555')
    years_fmt=dates.DateFormatter('%Y')
    ax.xaxis.set_major_formatter(years_fmt)
    
    for side in 'top,bottom,left,right'.split(','):
        ax.spines[side].set_visible(False)
        
    ax.set_facecolor('#1a1a1a')
    ax.tick_params(axis='both',length=0,which='minor')
        
    ax2=fig.add_axes((0.2,0.15,0.8,0.1))
    ax2.bar(df.date, [1]*len(df), width=31, color=cmap(df.HMI/df.HMI.max()), zorder=10)
    ax2.xaxis.set_major_locator(dates.YearLocator(base=10))
    ax2.xaxis.set_major_formatter(years_fmt)
    ax2.set_facecolor('#1a1a1a')
    for side in 'top,bottom,left,right'.split(','):
        ax2.spines[side].set_visible(False)
    ax2.tick_params(axis='y',length=0,labelleft=False)
    ax.set_title(' ', fontsize=15)
    ax.annotate('Housing Market Index',xy=(0,0), xytext=(.05,0.77),textcoords='figure fraction', xycoords='figure fraction', color='#ffffff',fontweight='bold', fontfamily='Roboto', fontsize=16)
    ax.annotate('Values >50 indicate positive sentiment',xy=(0.0,0.0), xytext=(.05,0.74),textcoords='figure fraction', xycoords='figure fraction', color='#ffffff', fontfamily='Roboto', fontsize=8)
    
    ax.axhline(y=5, color='#ffffff',zorder=12,linestyle='dashed')
    
    return fig
        
```

```python
(df
.stack()
.rename('hmi')
.reset_index()
.rename(columns=dict(zip(['   (Seasonally Adjusted)','level_1','hmi'],['Year','Month','HMI'])))
.assign(date=lambda df_:pd.to_datetime(df_.Month+'/'+df_.Year.astype(str)))
.pipe(plot_hmi)
)
```

```python
import pandas as pd
df = pd.read_csv("heart_failure_clinical_records_dataset.csv")
```

```python
df.head()
```

```python
print(df.isnull().sum())
```

```python
print(df.iloc[183:191,0:2])
df['age']=df['age'].astype(int)
```

```python
for column in df.columns:
    if column in ['anemia','diabetes','high_blood_pressure','sex','smoking','DEATH_EVENT']:
        print(column,": ",df[column].unique())
duplicates=df.duplicated().sum()
print("\nNumber of duplicates:",duplicates)
```

```python
df.describe()
```

```python
df.plot(y='creatinine_phosphokinase', use_index=True)
```

```python
df[['age','platelets','creatinine_phosphokinase','serum_creatinine','ejection_fraction','serum_sodium']].plot(kind='box',subplots=True,layout=(2,3))
```

```python
z=df[['anaemia','diabetes','high_blood_pressure','smoking','sex','DEATH_EVENT']].astype('category')
z.info()
```

```python
z.apply(lambda x: x.value_counts()).plot.bar()
```

```python
zz=z.apply(lambda x: x.value_counts())
print(zz)
```

```python
zzz=z[['anaemia','DEATH_EVENT']].apply(lambda x: x.value_counts())
print(list(zzz['DEATH_EVENT']))
zzz.plot(kind='bar')
```

```python
import numpy as np
zzz=z[['anaemia','DEATH_EVENT']]
zzz.pivot_table(index='anaemia',values=['DEATH_EVENT'],aggfunc=lambda x: len(x))
```

```python
import numpy as np
import matplotlib.pyplot as plt
zzz=z[['anaemia','DEATH_EVENT']]
yy=zzz.pivot_table(index=['anaemia'],columns=['DEATH_EVENT'],aggfunc=lambda x: len(x))
```

```python
z1=z[['high_blood_pressure','DEATH_EVENT']]
yyy=z1.pivot_table(index=['high_blood_pressure'],columns=['DEATH_EVENT'],aggfunc=lambda x: len(x))
```

```python
z2=z[['diabetes','DEATH_EVENT']]
yyyy=z2.pivot_table(index=['diabetes'],columns=['DEATH_EVENT'],aggfunc=lambda x: len(x))
```

```python
import pandas as pd
import numpy as np

# Creating a larger sample DataFrame with random data
np.random.seed(42)

# Generate random data for the example
num_rows = 1000
years = np.random.choice([2019, 2020, 2021], size=num_rows)
countries = np.random.choice(['USA', 'Canada', 'Mexico', 'Brazil', 'Argentina'], size=num_rows)
profits = np.random.randint(10000, 50000, size=num_rows)

sales_data = pd.DataFrame({'Year': years, 'Country': countries, 'Profit': profits})
sales_data
```

```python
# Using loc for subsetting
subset_loc = sales_data.loc[(sales_data['Year'] == 2020) & (sales_data['Country'].isin(['USA', 'Canada', 'Mexico'])) & (sales_data['Profit'] > 20000)]

# Using df[] for subsetting
subset_df = sales_data[(sales_data['Year'] == 2020) & (sales_data['Country'].isin(['USA', 'Canada', 'Mexico'])) & (sales_data['Profit'] > 20000)]
```

```python
display(subset_df)
```

```python
subset_query = sales_data.query('`Year` == 2020 and `Country` == ["USA", "Canada", "Mexico"] and `Profit` > 20000')
```

```python
display(subset_query)
```

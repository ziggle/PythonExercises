```python
import numpy as np
import scipy.stats as st
```

```python
#define sample data
data = [21.013,20.600,20.134,21.077,21.124,20.254]
```

```python
#create 95% confidence interval for population mean weight
m = np.mean(data)
print(f'mean = {np.mean(data)}')
x=st.t.interval(alpha=0.95, df=len(data)-1, loc=np.mean(data), scale=st.sem(data)) 
print(np.mean(data)-x[0],np.mean(data)-x[1])
```

```python
top = np.quantile(data,0.975)
bottom = np.quantile(data, 0.025)
print(f'top: {top=:.3f}, bottom: {bottom=:.3f}')
print(f'range interval: {top-m:.3f} to {bottom-m:.3f}')
```

```python
data2 = [19.509,19.127,18.676,19.559,19.626,18.795]
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
z = pd.DataFrame({'Vendor':data,'Starkey':data2}).melt(value_vars=['Vendor','Starkey'],value_name="Vrange")
print(z)
#sns.set_style('darkgrid')
plt.style.use(['fivethirtyeight','seaborn-deep'])
sns.displot(z,x="Vrange",kind='kde', fill=True,hue='variable')
plt.xlabel("Capacity (mAh)")
plt.title("Accu60 Capacity Distribution")
print(f'The Accu60 capacity distribution appears to be left skewed')
```

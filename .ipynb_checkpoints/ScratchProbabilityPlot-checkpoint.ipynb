```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import scipy.stats
import sklearn.linear_model
url="vehicles.csv.zip"
df = pd.read_csv(url)
df.head()
```

```python
# Create a probability plot

df2 =(df
      [['city08']]
      .sort_values(by='city08')
      .assign(
      quantiles=lambda df_:np.sort(np.random.normal(size=len(df))),
      )
)
```

```python
ax = (df2
     .plot.scatter(x='quantiles',y='city08',title='Probability Plot')
     )
lm=sklearn.linear_model.LinearRegression()
lm.fit(df2[['quantiles']],df2['city08'])

x = np.arange(-4,5)
plt.plot(x,lm.coef_*x+lm.intercept_,c='r')
```

```python
(auto
 .groupby(['Make','YEAR'])
 .Model
 .count()
 .unstack()
 .loc[['MITSUBISHI','FORD','NISSAN','TESLA','KIA'],[2013,2014,2015]]
 .plot.bar(width=0.7)
)
```

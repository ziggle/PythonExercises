```python
import numpy as np
import pandas as pd
df = pd.DataFrame(
    
    {"price": np.random.randn(1000) + 10},
    index=pd.date_range(start="2022-01-01", periods=1000, freq="D")
)
df = df.resample("10D").mean()
df.head()
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10,4))
plt.plot(df["price"])
plt.show()
```

```python
import matplotlib
matplotlib.__version__
```

```python
import matplotlib.dates as mdates
import matplotlib.ticker as ticker
plt.figure(figsize=(10,4))
plt.plot(df["price"])
# customize ticks on x-axis
plt.gca().xaxis.set_major_locator(mdates.MonthLocator(interval=2))
# rotate by 30 degrees
plt.xticks(rotation=30)
plt.show()
```

```python
plt.figure(figsize=(10,4))
plt.plot(df["price"])
# customize ticks on x-axis
plt.gca().xaxis.set_major_locator(mdates.MonthLocator(interval=2))
# rotate by 30 degrees
plt.xticks(rotation=30)
# customize ticks on y-axis
plt.gca().yaxis.set_major_locator(ticker.MultipleLocator(0.1))
plt.show()
```

```python
plt.figure(figsize=(10,4))
plt.plot(df["price"])
# customize ticks on x-axis
plt.gca().xaxis.set_major_locator(mdates.DayLocator(interval=90))
plt.gca().xaxis.set_major_formatter(
    mdates.DateFormatter('%d-%m-%Y')
)
# rotate by 30 degrees
plt.xticks(rotation=30)
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.plot(df["price"])
plt.gca().yaxis.set_major_locator(
   ticker.FixedLocator([9.2, 9.3, 9.4, 9.5, 10, 10.5])
)
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.plot(df["price"])
plt.gca().yaxis.set_major_locator(ticker.LinearLocator(numticks=10))
plt.show()
```

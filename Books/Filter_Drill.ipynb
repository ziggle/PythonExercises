```python

import pandas as pd
books=pd.read_csv("books.csv",header=1, thousands=r',',keep_default_na=False)
books.head()
```

```python
books.filter(["Author"])
```

```python
display(books['Author']=='Rowling, J. K')
```

```python

display(books[books['Author']=="Rowling, J. K."])
```

```python
display(books[(books['Volume'] < 1_000_000) & (books['Author']=="Rowling, J. K.")])
```

```python
books.query("Volume < 1_000_000 & Author=='Rowling, J. K.'")
```

```python
books.query("Author=='Rowling, J. K.' or Author=='Brown, Dan' or Author=='Meyer, Stephenie'")
```

```python
books[books.Author.isin(["Brown, Dan","Rowling, J. K.","Meyer, Stephenie"])]
```

```python
books[books.Author.str.contains("Rowl")]
```

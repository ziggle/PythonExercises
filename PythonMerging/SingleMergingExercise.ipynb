```python
import pandas as pd
df1 = pd.read_excel("Dummy_course_data.xlsx",
                    sheet_name="Fees")
df2 = pd.read_excel("Dummy_course_data.xlsx",
                    sheet_name="Discounts")
print(df1.head())
print(df2.head())
```

```python
df3=pd.merge(df1,df2)
print(df3)
```

```python
df4=pd.merge(df1, df2, how='outer')
print(df4)
```

```python
df5=pd.merge(df1, df2, how='left')
print(df5)
```

```python
df6=pd.merge(df1, df2, how='right')
print(df6)
```

```python
df7 = df1.join(df2,
               lsuffix = '_df1',
               rsuffix = '_df2')
df7
```

```python
df11 = df1.copy()
df11.set_index('Course', inplace=True)
df22 = df2.copy()
df22.set_index('Course', inplace=True)
df8 = df11.join(df22,
                lsuffix = '_df1',
                rsuffix = '_df2',
                on='Course')
df8
```

```python
df9=pd.concat([df1, df2])
print(df9)
```

```python
df10=pd.concat([df1,df2], axis=1)
df10
```

```python
df11=pd.concat([df1,df2], axis=1,
          keys = ["df1_data","df2_data"])
df11
```

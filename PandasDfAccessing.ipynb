```python
# Looks like Titanic dataset. I am just grabbing part of it for the drill.
import pandas as pd
import numpy as np
df=pd.DataFrame([
[0,0,3,"male",22,1,0,7.25,"S","Third","man",True,np.nan,"Southampton"],
[1,1,1,"female",38,1,0,71.28,"c","First","woman",False,"c","Cherbourg"],
[2,1,3,"female",26,0,0,7.92,"S","Third","woman",False,np.nan,"Southampton"],
[3,1,1,"female",35,1,0,53.1,"s","First","woman",False,"C","Queenstown"],
[4,0,3,"male",35,0,0,8.05,"S","Third","man",True,np.nan,"Southampton"],
[5,0,3,"male",12,1,0,7.25,"S","Third","man",True,np.nan,"Liverpool"],
[6,1,1,"female",13,1,0,71.28,"c","First","woman",False,"c","Cherbourg"],
[7,1,3,"female",14,0,0,7.92,"S","Third","woman",False,np.nan,"Liverpool"],
[8,1,1,"female",15,1,0,53.1,"s","First","woman",False,"C","Southampton"],
[9,0,3,"male",16,0,0,8.05,"S","Third","man",True,np.nan,"Southampton"],
],
columns=["Index","survived","pclass","sex","age","sibsp","parch",
         "fare embarked","Column1","class","who","adult_male_deck","Column2","embark_town"]
)
```

```python
display(df)
```

```python
# all rows and the "age" column:
df.loc[:, "age"]
```

```python
df.loc[:, ["age", "embark_town"]]
```

```python
# all rows and columns from "age" to "embark_town":
df.loc[:, "age":"embark_town"]
```

```python
# rows where the "age" is greater than 30 and all columns:
df.loc[df["age"] > 30, :]
```

```python
# rows where the "age" column is greater than 30 and only the "age" column:
df.loc[df["age"] > 30, "age"]
```

```python
# rows where the "embark_town" column is "Southampton", "Cherbourg", or "Queenstown", and all columns:
df.loc[df["embark_town"].isin(["Southampton", "Cherbourg", "Queenstown"]), :]
```

```python
# rows where the "age" column is greater than 30 and the "embark_town" column is "Southampton", and all columns:
df.loc[(df["age"] > 30) & (df["embark_town"] == "Southampton"), :]
```

```python
# all rows and the 3rd column ("age"):
df.iloc[:, 3]
```

```python
# all rows and the 3rd and 12th columns ("age" and "embark_town"):
df.iloc[:, [3, 12]]
```

```python
# all rows and columns from the 3rd to the 12th column (exclusive of the 12th column):
df.iloc[:, 3:12]
```

```python
# rows where the "embark_town" column is "Southampton", "Cherbourg", or "Queenstown", and all columns:

df.loc[df["embark_town"].isin(["Southampton", "Cherbourg", "Queenstown"]), :]
```

```python
# rows where the "age" column falls between 20 and 40 (inclusive), and all columns:
df.loc[df["age"].between(20, 40), :]
```

```python
# rows where the "class" column contains "Fir" and all columns:
df.loc[df["class"].str.contains("Fir"), :]
```

```python
# rows where the "class" column starts with "Fir" and all columns:
df.loc[df["class"].str.startswith("Fir"), :]
```

```python
# rows where the "class" column ends with "st" and all columns:
df.loc[df["class"].str.endswith("st"), :]
```

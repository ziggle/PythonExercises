```python
import matplotlib.pyplot as plt
fig=plt.figure(layout="constrained")
axes=fig.subplots(2,2)
axes[0,0].plot([1,2,3])
axes[0,1].scatter([1,2,3],[4,5,6])
axes[1,0].hist([1,2,3,4,5])
axes[1,1].bar([1,2,3],[2,4,6])
plt.show()
```

```python
import matplotlib.pyplot as plt
fig=plt.figure(layout="constrained")
ax_dict=fig.subplot_mosaic([['a','b'],['c','d']])
ax_dict['a'].plot([1,2,3])
ax_dict['b'].scatter([1,2,3],[4,5,6])
ax_dict['c'].hist([1,2,3,4,5])
ax_dict['d'].bar([1,2,3],[2,4,6])
plt.show()
```

```python
import matplotlib.pyplot as plt
fig=plt.figure(layout="constrained")
axes=fig.subplots(2,2)
mosaic="""
abd
ccd
"""
ax_dict=fig.subplot_mosaic(mosaic)
ax_dict['a'].plot([1,2,3])
ax_dict['b'].scatter([1,2,3],[4,5,6])
ax_dict['c'].hist([1,2,3,4,5])
ax_dict['d'].bar([1,2,3],[2,4,6])
```
